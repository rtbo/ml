```python
from scipy import datasets
import matplotlib.pyplot as plt

face = datasets.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
(h, w) = face.shape
face2 = face[ h//4 : -h//4 , w//4:-w//4 ]
plt.imshow(face2, cmap=plt.cm.gray)
plt.show()
255 /3
```

```python
face2[face2<85] = 0
face2[(face2>=85) & (face2 < 170)] = 128
face2[face2>192] = 255
plt.imshow(face2, cmap=plt.cm.gray)
plt.show()
```

## NaN corrections

```python
np.isnan(A).sum()/A.size
```
