## Exercice 14

```python
def graphique_14(data, **kwargs):
    plt.figure(figsize=kwargs.get('figsize', (8, len(data)*3.5)))
    for i, k in enumerate(data.keys()):
        plt.subplot(len(data), 1, i+1)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
dataset = {f'experience {i+1}': np.random.randn(100) for i in range(4)}

graphique_14(dataset)
```

## Exercice 17 - Pandas

Modifier les données du Titanic en categorisant les ages

```python
import pandas as pd

data = pd.read_excel("dataset/titanic.xls")
data = data.drop(['name', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest'], axis=1)
data = data.dropna(axis=0)
data.head()
```

```python
age = data['age']
agecat = age.copy().rename('agecat')
agecat[age < 20] = 0
agecat[(age >= 20) & (age < 30)] = 1
agecat[(age >= 30) & (age < 40)] = 2
agecat[age >= 40] = 3
data = pd.concat([data, agecat], axis=1)
data['agecat'].value_counts()
```
