```python
import matplotlib.pyplot as plt
import numpy as np

def graphique(dataset):
    plt.figure()
    for i, k in enumerate(dataset):
        plt.subplot(len(dataset), 1, i+1)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
```

```python
dataset = {f"experience {i+1}": np.random.randn(100) for i in range(4)}

graphique(dataset)
```
