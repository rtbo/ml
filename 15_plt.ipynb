```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris['data']
target = iris['target']
fnames = iris['feature_names']
tnames = iris['target_names']
```

## Scatter

```python
plt.scatter(data[:,0], data[:,1], c=target, alpha=0.5, s=data[:,2]*50)
plt.xlabel(fnames[0])
plt.ylabel(fnames[1])
```

## Graphique 3D

```python
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection="3d")
ax.scatter(data[:,0], data[:,1], data[:,2], c=target)
```

```python
f = lambda x, y: np.sin(x)+np.cos(x+y)*np.cos(x)

X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="plasma")
```

## Histogrammes

```python
plt.hist(data[:,0], bins=20)
plt.hist(data[:,1], bins=20)
```

```python
plt.hist2d(data[:,0], data[:,1], cmap="Blues")
plt.xlabel(fnames[0])
plt.ylabel(fnames[1])
plt.colorbar()
```

```python
import scipy.datasets as spd

face = spd.face(gray=True)
#plt.imshow(face, cmap="gray")
plt.hist(face.ravel(), bins=255)
plt.show()

```

## Contour plots

```python
plt.contour(X, Y, Z, 20)
```

```python
plt.contourf(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```

## Imshow

```python
plt.imshow(np.corrcoef(data.T), cmap="Blues")
plt.colorbar()
```
